```python
postcodes = %sql select pcd, county, country, electoral_division, local_authority_district, urban_rural_classification from postcode
```

```python
df = postcodes.DataFrame()
df
```

```python
dimensions = [
    'country', 
    'county', 
    'local_authority_district', 
    'electoral_division', 
    'urban_rural_classification',
]
```

```python
df['outcode'] = df['pcd'].str.split(' ').str[0]
df
```

```python
df[['outcode'] + dimensions].groupby('outcode').value_counts()
```

```python
df[['outcode'] + dimensions].groupby('outcode').value_counts().to_excel('test.xlsx')
```

```python
%%sql
select electoral_division, count(distinct local_authority_district), group_concat(distinct local_authority_district) as las
from postcode 
where electoral_division < 1575
group by electoral_division
order by count(distinct local_authority_district) DESC LIMIT 25
```
